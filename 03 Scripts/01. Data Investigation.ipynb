# 04. Cleaning Data / Minor Wrangling

## 04.1 Removing NaN values and creating new data frames for analyzing records that had NaN values in arrival and departure data


```python
# filling NaN values for tail_num with n/a 
# since a unique identifyier of airline code, flight number and day, month, year remain
df['TAIL_NUM'].fillna('NONE', inplace=True)
```

```python
# removal of Unnamed: 21 column, OP_CARRIER_AIRLINE_ID, ORIGIN_AIRPORT_ID, ORIGIN_AIRPORT_SEQ_ID, DEST_AIRPORT_ID, DEST_AIRPORT_SEQ_ID columns
df = df.drop(['Unnamed: 21','OP_CARRIER_AIRLINE_ID', 'ORIGIN_AIRPORT_ID', 'ORIGIN_AIRPORT_SEQ_ID', 'DEST_AIRPORT_ID', 'DEST_AIRPORT_SEQ_ID'], axis=1)
```

```python
# adding in location data based on airport data starting with dictionary of airport and geography = ad for airport dictionary
ad
```

```python
sd = sd.drop(['Unnamed: 2'], axis=1)
```

```python
ad = pd.merge(ad, sd, on='STATE_OR_REGION', how='left')
```

```python
ad
```

```python
# adding in the geographic data for destination airport
df = pd.merge(df, ad[['ORIGIN','CITY','STATE_OR_REGION','FULL_NAME','GMT']], on='ORIGIN', how='left')
```

```python
# adding in the geographic data for destination airport
df = pd.merge(df, ad[['DEST','CITY','STATE_OR_REGION','FULL_NAME','GMT']], on='DEST', how='left')
```

```python
# renaming city and origin
df.rename(columns={'CITY_x':'ORIGIN_CITY','CITY_y':'DEST_CITY','STATE_OR_REGION_x':'ORIGIN_STATE_OR_REGION','STATE_OR_REGION_y':'DEST_STATE_OR_REGION','FULL_NAME_x':'ORIGIN_FULL_NAME','FULL_NAME_y':'DEST_FULL_NAME', 'GMT_x': 'GMT_DEP', 'GMT_y': 'GMT_ARR'}, inplace=True)
```

```python
df
```

```python
# succesful, canceled, diverted flights data frame to compare with removal of time based columns due to their NaN values
scd = df.drop(['ARR_TIME','ARR_DEL15','DEP_TIME','DEP_DEL15'], axis=1)
```

```python
scd.isnull().sum()
```

```python
# confirming removal of column
df.info()
```

```python
# investigating if cancelled records have NaN 
cancelled = df[df['CANCELLED'] == 1]
```

```python
# review of null values in CANCELLED flights
cancelled.isnull().sum()
```

```python
cancelled = cancelled.drop(['ARR_TIME','ARR_DEL15','DEP_TIME','DEP_DEL15'], axis=1)
```

```python
# confirming removal
cancelled.isnull().sum()
```

```python
# investigating remaining null values
df[df['ARR_DEL15'].isnull()]
```

```python
# saw a number of NaN values were diverted, investigating if they are the remaining NaN records
diverted = df[df['DIVERTED'] == 1]
```

```python
# all remaining null values accounted for
diverted.isnull().sum()
```

```python
# diverted has good departure time data so in comparing with all other flights I am only removing columns with NaN values
diverted = df.drop(['ARR_TIME','ARR_DEL15'], axis=1)
```

```python
# removing all cancelled flight records from diverted data frame due to the NaN values associated with the grouping
diverted = diverted[diverted['CANCELLED'] == 0]
```

```python
# removing all cancelled flight records from main data frame
df = df[df['CANCELLED'] == 0]
```

```python
# removing all diverted flight records from main data frame
df = df[df['DIVERTED'] == 0]
```

```python
# removing canceled and diverted columns since they are now data sets
df = df.drop(['DIVERTED', 'CANCELLED'], axis=1)
```

## 04.2 Wrangling: changing from object to category to save space.

```python
# Creating list of columns and what data type to change to
new_data_types = {
    'OP_UNIQUE_CARRIER': 'category',
    'OP_CARRIER': 'category',
    'TAIL_NUM': 'category',
    'ORIGIN': 'category',
    'DEST': 'category',
    'DEP_TIME_BLK': 'category'    
}
```

```python
# Changing data type for list of columns
df = df.astype(new_data_types)
```
